```python
import pandas as pd

df = pd.read_csv('chromo.txt', sep='\t')
a=[]
for i in range (12000):
    if df['Time (min)'][i]>=4 and df['Time (min)'][i]<=7:
        a.append(i)

df=pd.read_csv('chromo.txt', sep='\t', skiprows=min(a), skipfooter=12000-max(a), names=['Time (min)', 'Value (mAU)'])

df.plot(df.columns[0], df.columns[1], ylabel=df.columns[1], title = 'chromo (4-7 min)')
```
