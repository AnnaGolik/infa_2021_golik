```python
l = int(input())
a=[]
for i in range(l):
    a.append(int(input()))
if l%2==0:
    print("медиана: " + str(int(a[l//2-1]+a[l//2])/2))
else:
    print("медиана: " + str(int((a[(l-1)//2-1]+a[(l+1)//2])/2)))
```

```python
import math
l = int(input())
a=[]
for i in range(l):
    a.append(int(input()))
m=int(math.fsum(a)//l)
b=[]
for i in range (l):
    b.append((a[i]-m)**2)
s=int(math.fsum(b))
d=(s/(l))**0.5
print("среднеквадратичное отклонение: "+ str(d))

def med(a,l):
    if l%2==0:
        med=int(a[l//2-1]+a[l//2])/2
    else:
        med=int((a[(l-1)//2-1]+a[(l+1)//2])/2)
    return med

a1=[]
a2=[]
for i in range (int(l/2)):
    a1.append(a[i])
    a2.append(a[int(l-i-1)])

med1=med(a1, len(a1))
med2=med(a2, len(a2))
r=med2-med1
print('межквартильный размах: '+ str(r))
```

```python
import matplotlib.pyplot as plt
colors = ['blue', 'orange', 'green', 'red', 'violet']
labels = ['А.', 'Б.', 'В.','Г.', 'Д.']
x = [1, 2, 3, 4, 5]
y1 = [16, 18, 20, 22, 24]
y2 = [20, 20, 18, 22, 20]
y3= [24, 22, 20, 18, 16]

plt.title('За месяц', size=14)
plt.bar(x, y1, color=colors, edgecolor='k', tick_label=labels)
plt.grid(True, axis='y')
plt.show()

plt.title('За неделю', size=14)
plt.bar(x, y2, color=colors, edgecolor='k', tick_label=labels)
plt.grid(True, axis='y')
plt.show()

plt.title('Экзит-полл', size=14)
plt.bar(x, y3, color=colors, edgecolor='k', tick_label=labels)
plt.grid(True, axis='y')
plt.show()

data1 = y1
plt.title('За месяц', size=14)
plt.pie(data1, labels=('А.', 'Б.', 'В.','Г.', 'Д.'))
plt.show()

data2 = y2
plt.title('За неделю', size=14)
plt.pie(data2, labels=('А.', 'Б.', 'В.','Г.', 'Д.'))
plt.show()

data3 = y3
plt.title('Экзит-полл', size=14)
plt.pie(data3, labels=('А.', 'Б.', 'В.','Г.', 'Д.'))
plt.show()

#столбчатые диаграммы более наглядные
```
