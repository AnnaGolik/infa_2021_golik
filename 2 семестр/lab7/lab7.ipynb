```python
import pandas as pd
import matplotlib.pyplot as plt

students = pd.read_csv("StudentsPerformance.csv")

bd = students[students["parental level of education"]=="bachelor's degree"] 
sc = students[students["parental level of education"]=='some college'] 
md = students[students["parental level of education"]=="master's degree"] 
ad = students[students["parental level of education"]=="associate's degree"] 
hs = students[students["parental level of education"]=='high school'] 
shs = students[students["parental level of education"]=='some high school'] 


for df in (bd, sc, md, ad, hs, shs):
    plt.scatter(df["reading score"], df["math score"])
parental_levelof_ed_types = ["bachelors\ndegree", "some\ncollege", 
                             'masters\ndegree', 'associates\ndegree', 'high\nschool', 'some\nhigh\nschool']
plt.legend(parental_levelof_ed_types)
plt.title('Parental level of education,\nReading and Math score')
plt.xlabel('Reading score')
plt.ylabel('Math score')
plt.show()

for df in (bd, sc, md, ad, hs, shs):
    plt.scatter(df["reading score"], df["writing score"])
parental_levelof_ed_types = ["bachelors\ndegree", "some\ncollege", 
                             'masters\ndegree', 'associates\ndegree', 'high\nschool', 'some\nhigh\nschool']
plt.legend(parental_levelof_ed_types)
plt.title('Parental level of education,\nReading and Writing score')
plt.xlabel('Reading score')
plt.ylabel('Writing score')
plt.show()

types=["reading score", "math score", "writing score"]
for i in range (3):
    plt.boxplot((bd[types[i]], sc[types[i]], md[types[i]], ad[types[i]],
                 hs[types[i]], shs[types[i]]),
                labels=parental_levelof_ed_types)
    plt.title('Parental level of education,\n'+types[i])
    plt.xlabel('parental level of education')
    plt.ylabel(types[i])
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

mice = pd.read_csv('miceprotein.csv', na_values='?')

mice_classes=[*mice['class'].unique()]

levels_of_exp=[]
for i in range (len(mice_classes)):
    levels_of_exp.append(mice[mice['class']==mice_classes[i]])
    
plt.figure(figsize=(15, 7))
for i in range (len(levels_of_exp)):
    sp=plt.subplot(1,len(levels_of_exp), i+1)
    sp.set_ylim([min(mice['DYRK1A_N']), max(mice['DYRK1A_N'])])
    sp.boxplot(levels_of_exp[i]['DYRK1A_N'].dropna(), labels=[mice_classes[i]])
    plt.title('DYRK1A_N,\n'+mice_classes[i])
plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 7))
for i in range (len(levels_of_exp)):
    sp=plt.subplot(1,len(levels_of_exp), i+1)
    sp.set_ylim([min(mice['SOD1_N'].dropna()), max(mice['SOD1_N'].dropna())])
    sp.boxplot(levels_of_exp[i]['SOD1_N'].dropna(), labels=[mice_classes[i]])
    plt.title('SOD1_N,\n'+mice_classes[i])
plt.tight_layout()
plt.show()
```
