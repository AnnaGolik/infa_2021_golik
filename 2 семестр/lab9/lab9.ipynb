```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

students = pd.read_csv("StudentsPerformance.csv")

first = students[students["test preparation course"]=="none"] 
second = students[students["test preparation course"]=="completed"] 

for type in ('reading score', 'math score', 'writing score'):
    colors=['blue', 'red']
    x = ['none', 'completed']
    hh=[]
    yy=[]
    for df in (first, second):
        h = np.mean(df[type]) #высота столбца = среднее значение
        y=np.std(df[type]) #погрешность = стандартное оклонение
        hh.append(h)
        yy.append(y)
    plt.bar(x, height=hh, yerr=yy, color = colors)
    plt.xlabel(type)
    plt.show()
```

```python
import pandas as pd

df = pd.read_csv('chromo.txt', sep='\t')
a=[]
for i in range (df.shape[0]):
    if df['Time (min)'][i]>=4 and df['Time (min)'][i]<=7:
        a.append(i)

df=pd.read_csv('chromo.txt', sep='\t', skiprows=min(a), skipfooter=df.shape[0]-max(a), names=['Time (min)', 'Value (mAU)'])

maxel=max(df[df.columns[1]]) #максимальное значение
ind = df[df[df.columns[1]] == maxel].index.values.astype(int) #индекс максимального значения
xmean=df[df.columns[0]][int(ind)] #координата по оси абцисс максимального значения
       
df.plot(df.columns[0], df.columns[1], ylabel=df.columns[1], title = 'chromo (4-7 min)')
plt.arrow(x=xmean, y=20, dx=0, dy=-(20-maxel-0.2), width = 0.03,  head_width = 0.1)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

students = pd.read_csv("StudentsPerformance.csv")

numbers=[50,66,85]

linecolours = ['purple', 'green', 'yellow']
type = ('reading score', 'math score', 'writing score')
plt.figure(figsize=(30,15))
plt.scatter(students[type[0]], students[type[1]], c=students[type[2]])
for i in range (3):
    plt.hlines(numbers[i], 0, 100, colors=linecolours[i])
    plt.vlines(numbers[i], 0, 100, colors=linecolours[i], label = numbers[i])
    plt.legend()
plt.show() 
```
