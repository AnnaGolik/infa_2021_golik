```python
import pandas as pd
df = pd.read_excel('task2_column_names.ods')

cols=df.columns
colnames=[]
for i in range(len(cols)):
    colnames.append(cols[i])
print(colnames)

names=[]
for i in range (len(colnames)):
    
    result = ''

    if colnames[i][0] == 't':
        result='t'
        names.append(result)
        continue
        
    if colnames[i][0] == 'A':
        result = ' ,'
    elif colnames[i][0] == 'S':
        result = ' ,'

    a=colnames[i][1:-1]
    result=result+'   '+ a + ' '
        
    if colnames[i][-1] == '+':
        result = result + ' ( )'
    elif colnames[i][-1] == '-':
        result = result + ' ( )'
    
    names.append(result)

df = pd.read_excel('task2_column_names.ods', names=names)
df

filename = 'task2_renamed.xls'
df.to_excel(filename)
```
